```python
import re
```

```python
class Calculator:
    def __init__(self, s):
        self.s = s
    
    @staticmethod
    def compute(s):
        s = s.replace(" ", "")
        if s[0] == '+':
            s = s[1:]
            
        nums = re.split("[\\+\\-\\*\\/]", s)
        ops = re.findall("[\\+\\-\\*\\/]", s)
       
        for n, i in enumerate(ops):
            if (i == '+') | (i == '-'):
                nums[n + 1] = i + nums[n + 1]
            if i == "-":
                ops[n] = '+'
                
        if s[0] == '-':
            del(nums[0])
            del(ops[0])
            
        #print(nums,ops)
        while len(nums)  > 1:
            for n, o in enumerate(ops):
                if o == '/':
                    nums[n] = float(nums[n]) / float(nums[n+1])
                    del(nums[n+1])
                    del(ops[n])
                    break

                if o == '*':
                    if ops.count("/") == 0:
                        nums[n] = float(nums[n]) * float(nums[n+1])
                        del(nums[n+1])
                        del(ops[n])
                        break

                if o == '+':
                    if (ops.count("/") == 0) & (ops.count("*") == 0):
                        nums[n] = float(nums[n]) + float(nums[n+1])
                        del(nums[n+1])
                        del(ops[n])
                        break

                if o == '-':
                    if (ops.count("/") == 0) & (ops.count("*") == 0) & (ops.count("+") == 0):
                        nums[n] = float(nums[n]) - float(nums[n+1])
                        del(nums[n+1])
                        del(ops[n])
                        break
            #print(nums,ops)
    
        return nums[0]
    
    def __repr__(self):
        return str(self.compute(self.s))
```
