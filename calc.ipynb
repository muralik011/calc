```python
class Calculator:
    def __init__(self, s):
        self.s = s
    
    def compute(self):
        if self.s.find('+') >= 0:
            a, b = self.s.split('+')
            return float(a) + float(b)
        elif self.s.find('-') >= 0:
            a, b = self.s.split('-')
            return float(a) - float(b)
        elif self.s.find('/') >= 0:
            a, b = self.s.split('/')
            return float(a) / float(b)
        elif self.s.find('*') >= 0:
            a, b = self.s.split('*')
            return float(a) * float(b)
        else:
            return "Incorrect input"
        
    def __repr__(self):
        return str(self.compute())
```

```python
Calculator("2+3")
```

```python
Calculator("2-3")
```

```python
Calculator("2*3")
```

```python
Calculator("2/3")
```
